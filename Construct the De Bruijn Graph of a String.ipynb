De Bruijn graph of m symbols is a directed graph representing overlaps between sequences of symbols. It has m^{n} vertices, consisting of all possible length-n sequences of the given symbols; the same symbol may appear multiple times in a sequence.
So what important in this De Bruijn graph is that we will construct node based on composition. The output will be the generated node or path.

Example this is the composition AAGATTCTCTAC
so the node that will be generated is for each kmers is 
AAG -> AGA -> GAT -> ATT -> TTC and so on
